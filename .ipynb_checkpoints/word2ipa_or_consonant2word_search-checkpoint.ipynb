```python
import sys
import requests
import json
import vlc
import re
import random
from IPython.display import Audio

def load_json_database(url):
    records = []
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad status codes
        for line in response.iter_lines(decode_unicode=True):
            if line:
                try:
                    record = json.loads(line)
                    records.append(record)
                except json.JSONDecodeError as e:
                    print(f"Error parsing JSON: {e}")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from URL: {e}")
    return records

def search_in_json_database(database, search_word, region):
    for record in database:
        # 检查 word 字段是否匹配
        if record.get('word') == search_word:
            # 找到匹配项后，获取美式发音信息
            pos_items = record.get('pos_items', [])
            for pos_item in pos_items:
                pronunciations = pos_item.get('pronunciations', [])
                for pronunciation in pronunciations:
                    if pronunciation.get('region') == region:
                        # 找到美式发音，返回相关信息
                        return {
                            'pronunciation': pronunciation.get('pronunciation'),
                            'audio': pronunciation.get('audio')
                        }
    # 如果没有找到匹配的 word 字段，返回 'not exist'
    return 'not exist'

url = "https://raw.githubusercontent.com/zelic91/camdict/main/cam_dict.refined.json"

json_database = load_json_database(url)
```

```python
import requests
import json
import re
import random

# 假设你已经定义了这个函数
def search_in_json_database(word):
    """
    在 JSON 格式的数据库中搜索单词,返回发音和音频 URL.
    """
    # 假设你的数据库是一个 URL
    database_url = "https://raw.githubusercontent.com/zelic91/camdict/main/cam_dict.refined.json"
    
    # 从 URL 获取数据库
    response = requests.get(database_url)
    database = json.loads(response.text)
    
    # 在数据库中搜索单词
    for entry in database:
        if entry["word"].lower() == word.lower():
            pronunciation = entry
            return pronunciation
    
    # 如果单词没有找到,返回 None
    return None

# 使用示例
word = "hello"
result = search_in_json_database(word)
if result:
    print(f"Pronunciation: {result['pronunciation']}")
    print(f"Audio URL: {result['audio']}")
else:
    print(f"Sorry, the word '{word}' was not found in the database.")
```

```python
# 通过单词，查询对应的美式英语音标和英式英语音标。
import re
import json

def extract_pronunciation(word):
    with open('/Users/chengyong/Downloads/cam_dict.refined.json', 'r') as f:
        for line in f:
            try:
                entry = json.loads(line)
                if entry['word'] == word:
                    pronunciations = entry['pos_items'][0]['pronunciations']
                    for pronunciation in pronunciations:
                        region = pronunciation['region']
                        ipa = pronunciation['pronunciation']
                        print(f"Region: {region}, Pronunciation: {ipa}")
                    return
            except json.JSONDecodeError:
                continue

    print(f"No pronunciation found for '{word}'")

# 使用示例
extract_pronunciation('aardvarks')
extract_pronunciation('aah')
extract_pronunciation('aback')
```

```python
# 通过特定辅音，找包含这个辅音的所有单词，输出为 markdown。
import json
import re

def find_words_with_consonant(json_file, consonant):
    """
    查找 JSON 文档中含有特定辅音的单词。
    
    参数:
    json_file (str): JSON 文件的路径
    consonant (str): 要查找的辅音
    """
    results = []
    with open(json_file, 'r') as f:
        for line in f:
            try:
                item = json.loads(line.strip())
                word = item['word']
                pronunciations = [p['pronunciation'] for pos in item['pos_items'] for p in pos['pronunciations']]
                
                for pronunciation in pronunciations:
                    if re.search(rf'\b\w*[{consonant}]\w*\b', pronunciation):
                        results.append((word, pronunciation))
            except (KeyError, ValueError):
                # 跳过格式错误的行
                continue
    
    # 输出 Markdown 表格
    print("| Word | Pronunciation |")
    print("| --- | --- |")
    for word, pronunciation in results:
        print(f"| {word} | {pronunciation} |")


# 使用示例，特殊的辅音如ɚ，t̬
find_words_with_consonant('/Users/chengyong/Downloads/cam_dict.refined.json', 't̬')
```
